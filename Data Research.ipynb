### Arts Council National Lottery Project Grants awards made between 01 April 2022 - 31st March 2023

### All the lists provide the following information about the awards:

- **Recipient** _The name of the individual or organisation in receipt of an award_
- **Activity name** The name given to the activity by the recipient
- **Award amount** Amount of money awarded to the recipient. This is subject to change, and are correct as of the completed date given above.
- **Decision date** The date of the decision by the Arts Council to offer the award. Actual activity will be later this date, and may not commence in the same financial year as the decision.
- **Decision month** The calendar month in which the award offer was made
- **Decision quarter** The quarter of the financial year in which the award offer was made
- **ACE Area** The assigned Arts Council Area for that project. This is in relation to where the activity takes place and the area budget funding that activity. This may be different to other location data which is based on administrative postcode.
- **Local authority** UK Local authority district in which the recipient resides. This is based on the applicants postcode, and therefore does not always reflect where the funded activity took place. This field may show as 0 if the applicant's postcode is outside of the UK.
- **Main discipline** The main type of activity the funded activity involves as identified by the Arts Council
- **Time-Limited Priority** Specific types of applications that we prioritise for a time-limited period to help us meet our Strategy. This includes Supporting Grassroots Live Music, which was defined as a ‘strand’ in previous reporting years’

Please note: Awards that were offered but subsequently withdrawn are not included in these lists

```python
category_names = ['Strongly disagree', 'Disagree',
                  'Neither agree nor disagree', 'Agree', 'Strongly agree']
results = {
    'Question 1': [10, 15, 17, 32, 26],
    'Question 2': [26, 22, 29, 10, 13],
    'Question 3': [35, 37, 7, 2, 19],
    'Question 4': [32, 11, 9, 15, 33],
    'Question 5': [21, 29, 5, 5, 40],
    'Question 6': [8, 19, 5, 30, 38]
}


def survey(results, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncols=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax


survey(results, category_names)
plt.show()
```
